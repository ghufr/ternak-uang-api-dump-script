# Scraping Data - Ternak Uang

```python
import requests
import json
```

```python
BASE_URL = "https://api.ternakuang.id/api/"
```

# Events
There's no guarantee that you'll granted entry to zoom meeting

```python
events = json.loads(requests.get(BASE_URL + "events").text)

for event in events["data"]:
    title = event["title"]
    event_url = event["event_url"]
    
    print(f"{title} - {event_url}")
    
# print(json.dumps(events["data"], indent=2, sort_keys=True))
```

# Reports

```python
reports = json.loads(requests.get(BASE_URL + "reports?page=1&limit=200").text)

for report in reports["data"]:
    title = report["title"]
    pdf_url = report["thumbnail_url"].replace("tn_", "").replace(".png", ".pdf").replace("__", "_")
    
    print(f"{title}: {pdf_url}")
```
